# Correlation Error Plots (CNOISE)

### Install Dependencies

```r
install.packages("corrplot")
```

### Use Dependencies and load datasets

```r
library(corrplot)
source("http://www.sthda.com/upload/rquery_cormat.r")

# Read datasets
adaptive <- read.delim("cnoise/adaptive.tsv", row.names=NULL)
fixed <- read.delim("cnoise/fixed.tsv", row.names=NULL)
selective <- read.delim("cnoise/selective.tsv", row.names=NULL)
```

## Correlation Matrices
### Adaptive

```r
# get values where line = 1
adaptive_line1_nodes_only <- subset(adaptive, lines == 1)#process data into long type
# remove line column
adaptive_line1_nodes_only$lines <- NULL
cormat <- rquery.cormat(adaptive_line1_nodes_only, type = "upper")
```

### Fixed
Note that some of the XYERR values are NA

```r
# get values where line = 1
selective_line1_nodes_only <- subset(selective, lines == 1)#process data into long type
# remove line column
selective_line1_nodes_only$lines <- NULL
cormat <- rquery.cormat(selective_line1_nodes_only, type = "upper")
```
