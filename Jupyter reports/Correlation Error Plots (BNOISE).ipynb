# Correlation Error Plots (BNOISE)

### Install Dependencies

```r
install.packages("corrplot")
```

### Use Dependencies and load datasets

```r
library(corrplot)
source("http://www.sthda.com/upload/rquery_cormat.r")

# Read datasets
adaptive <- read.delim("bnoise/adaptive.txt", row.names=NULL)
fixed <- read.delim("bnoise/fixed.txt", row.names=NULL)
selective <- read.delim("bnoise/selective.txt", row.names=NULL)
```

## Correlation Matrices
### Adaptive

```r
cormat <- rquery.cormat(adaptive, type = "upper")
adaptive
```

### Fixed
Note that some of the XYERR values are NA

```r
cormat <- rquery.cormat(fixed, type = "upper")
```

### Selective

```r
    cormat <- rquery.cormat(selective, type = "upper")
```
